```python
# Calculadora em Python com menu interativo e repetição até o usuário sair

while True:
    print("\n=== CALCULADORA ===")
    print("Escolha a operação:")
    print("1. Soma (+)")
    print("2. Subtração (-)")
    print("3. Multiplicação (*)")
    print("4. Divisão (/)")
    print("5. Potência (**)")
    print("6. Sair")

    escolha = input("Digite o número da operação desejada: ")

    if escolha == "6":
        print("Saindo da calculadora. Até mais!")
        break

    # Solicita os dois números
    try:
        numero1 = float(input("Digite o primeiro número: "))
        numero2 = float(input("Digite o segundo número: "))
    except ValueError:
        print("Erro: por favor, digite apenas números válidos.")
        continue

    # Realiza a operação com base na escolha
    if escolha == "1":
        print(f"Resultado da soma: {numero1 + numero2}")
    elif escolha == "2":
        print(f"Resultado da subtração: {numero1 - numero2}")
    elif escolha == "3":
        print(f"Resultado da multiplicação: {numero1 * numero2}")
    elif escolha == "4":
        if numero2 != 0:
            print(f"Resultado da divisão: {numero1 / numero2}")
        else:
            print("Erro: divisão por zero!")
    elif escolha == "5":
        print(f"Resultado da potência: {numero1 ** numero2}")
    else:
        print("Opção inválida. Por favor, escolha uma das opções do menu.")
```
